## MINST Dataset

### import the libraries

```python
# ## import the libraries..
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# import the MNIST dataset from sklearn.datasets.load_disgits 
from sklearn.datasets import load_digits
```

```python
## Store the dataset which is used further.
digits=load_digits()
```

```python
# To check the shape of dataset use.shape()
print(digits.data.shape)
```

```python
# plot the data onto graph
plt.gray() 
plt.matshow(digits.images[0]) 
plt.show()
```

```python
# By slicing we get the two rows but our data is np.array form so this show mutidimensional array
digits.data[:2]
```

```python

# Strore the features of Mnist dataset into X
X=np.array(digits.data)
```

```python

# Store the target Mnist values into Y
y=np.array(digits.target)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=10,random_state=0)
clusters=kmeans.fit_predict(digits.data)
```

```python

print clusters
```

```python
digits.data.shape
```

```python
kmeans.cluster_centers_ .shape
```

```python
from scipy.stats import mode
labels=np.zeros_like(clusters)

for i in range(10):
    a=(clusters==i)

#     print a
#     print  (digits.target[a])[0]
    labels[a]=mode(digits.target[a])[0]
```

```python
labels.shape
```

## Accuracy

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(digits.target,labels)
```

## Confusion Matrix

```python
confusion_matrix(digits.target,labels)
```

## IMAGE

```python
from sklearn.datasets import load_sample_image
```

```python
flower=load_sample_image('flower.jpg')
```

```python

type(flower)
```

```python
# plt.axes(xticks=[],yticks=[])
plt.imshow(flower)
```

```python
print flower.shape
```

```python
flower
```

```python
flower=flower/255.0
```

```python
image_array=flower.reshape(flower.shape[0]*flower.shape[1],flower.shape[2])
```

```python

image_array.shape
```

```python
image_array
```

```python
from sklearn.utils import shuffle
image_array_sample = shuffle(image_array, random_state=0)[:6000]
```

```python
kimage=KMeans(n_clusters=7,random_state=0).fit(image_array_sample)
```

```python
labels = kimage.predict(image_array)
```

## Original image

```python
plt.figure(1)
plt.clf()
ax = plt.axes([0, 0, 1, 1])
plt.axis('off')
plt.title('Original image ')
plt.imshow(flower)
```

```python
new_colors=kimage.cluster_centers_[labels]
```

```python
flower_color=new_colors.reshape(flower.shape)
```

## Quantized image with 7 clusters

```python
plt.figure(2)
plt.clf()
ax = plt.axes([0, 0, 1, 1])
plt.axis('off')
plt.title('Quantized image (7 colors, K-Means)')
plt.imshow(flower_color)
```
